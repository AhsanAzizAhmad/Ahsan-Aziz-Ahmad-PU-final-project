```python
import matplotlib.pyplot as plt
import seaborn as sns
```

### Here is the dataset for poverty from 2000 to 2023.We will see which factors affecting poverty

### 1) HDI ( international recognized poverty indicator) 

```python
# Setting style
sns.set(style="whitegrid")

# Plot poverty over the years
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Year', y='Poverty Headcount Ratio (%)', marker='o', color='teal')
plt.title('Poverty Headcount Ratio in Pakistan (2000-2023)', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Poverty Headcount Ratio (%)', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Poverty Trend : Poverty mostly decreased but rose after 2019

### 2) Poverty vs Growth trend

```python
# Poverty vs GDP Growth scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='GDP Growth Rate (%)', y='Poverty Headcount Ratio (%)', hue='Year', palette='viridis', s=100)
plt.title('Poverty vs GDP Growth Rate (2000-2023)', fontsize=16)
plt.xlabel('GDP Growth Rate (%)', fontsize=14)
plt.ylabel('Poverty Headcount Ratio (%)', fontsize=14)
plt.legend(title='Year', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# Calculate the correlation
correlation = df['GDP Growth Rate (%)'].corr(df['Poverty Headcount Ratio (%)'])
print(f"Correlation between GDP Growth and Poverty Rate: {correlation:.2f}")
```

### There is negative correlation (higher GDP growth leads to lower poverty)

### 3) Poverty vs Inflation rate 

```python
# Poverty vs Inflation scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Inflation Rate (%)', y='Poverty Headcount Ratio (%)', hue='Year', palette='coolwarm', s=100)
plt.title('Poverty vs Inflation Rate (2000-2023)', fontsize=16)
plt.xlabel('Inflation Rate (%)', fontsize=14)
plt.ylabel('Poverty Headcount Ratio (%)', fontsize=14)
plt.legend(title='Year', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# Calculate the correlation
correlation_inflation = df['Inflation Rate (%)'].corr(df['Poverty Headcount Ratio (%)'])
print(f"Correlation between Inflation Rate and Poverty Rate: {correlation_inflation:.2f}")
```

### positive correlation (higher inflation leads to higher poverty)

### 4) Poverty vs Government Social Spending

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Government Social Spending (% of GDP)', y='Poverty Headcount Ratio (%)', hue='Year', palette='Spectral', s=100)
plt.title('Poverty vs Government Social Spending (2000-2023)', fontsize=16)
plt.xlabel('Government Social Spending (% of GDP)', fontsize=14)
plt.ylabel('Poverty Headcount Ratio (%)', fontsize=14)
plt.legend(title='Year', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# Calculate correlation
correlation_spending = df['Government Social Spending (% of GDP)'].corr(df['Poverty Headcount Ratio (%)'])
print(f"Correlation between Social Spending and Poverty Rate: {correlation_spending:.2f}")
```

### negative correlation (more spending leads to lower poverty)

### 5) Poverty vs Climate Disasters

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Climate Disasters (count)', y='Poverty Headcount Ratio (%)', hue='Year', palette='plasma', s=100)
plt.title('Poverty vs Climate Disasters (2000-2023)', fontsize=16)
plt.xlabel('Climate Disasters (count)', fontsize=14)
plt.ylabel('Poverty Headcount Ratio (%)', fontsize=14)
plt.legend(title='Year', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# Calculate correlation
correlation_disasters = df['Climate Disasters (count)'].corr(df['Poverty Headcount Ratio (%)'])
print(f"Correlation between Climate Disasters and Poverty Rate: {correlation_disasters:.2f}")
```

### Positive corelation means more climate disasters leads to more poverty
